–í—ñ–¥–ø–æ–≤—ñ–¥—å: –º–∞–π–∂–µ 52% –Ω–∞—à–∏—Ö –ø—Ä–æ–¥–∞–∂—ñ–≤ –≤—ñ–¥ –∂—ñ–Ω–æ–∫, 48% –≤—ñ–¥ —á–æ–ª–æ–≤—ñ–∫—ñ–≤

```python
Total_by_gender = df.groupby(['Gender'])['Total'].sum()

Gender_structure = (Total_by_gender / Total_by_gender.sum()) * 100

Gender_structure
```

–í—ñ–¥–ø–æ–≤—ñ–¥—å: –∂—ñ–Ω–∫–∏ –Ω–∞–π–±—ñ–ª—å—à–µ –∫—É–ø—É—é—Ç—å Food and beverages (19.76%)	Fashion accessories (18.13%)	Home and lifestyle	(17.89%). 
—á–æ–ª–æ–≤—ñ–∫–∏ –Ω–∞–π–±—ñ–ª—å—à–µ –∫—É–ø—É—é—Ç—å Health and beauty (19.75%)	Electronic accessories (17.56%)	Sports and travel	(17.12%).

```python
table = df.pivot_table(index='Product line', columns='Gender',values='Total',aggfunc='sum')
percent_table = table / table.sum() * 100

percent_table.round(2)
```
